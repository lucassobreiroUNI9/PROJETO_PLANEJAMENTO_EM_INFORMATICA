```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

ocorrencia = pd.read_csv('/content/ocorrencias.csv')

classificacao = ocorrencia.groupby('data')['ano'].sum().plot(grid=True, figsize=(10,5), color='#0D47A1', alpha=.7)

for p in classificacao.patches:
    b=p.get_bbox()
    classificacao.annotate("{:.0f}".format(b.y1 + b.y0), ((b.x0 + b.x1)/2 - 0.03, b.y1 + 15))

plt.xticks(rotation=0)
plt.xlabel('CLASSIFICAÇÃO DA OCORRÊNCIA')
plt.ylabel('QUANTIDADE')
plt.title('QUANTIDADE DE OCORRÊNCIAS POR CLASSIFICAÇÃO \n Total: ' + str(ocorrencia['ano'].sum()) + ' ocorrências')
plt.show()
```
